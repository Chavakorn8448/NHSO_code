```python
import numpy as np
from tqdm.auto import tqdm
import torch
from functools import partial

#transformers
from transformers import (
    CamembertTokenizer,
    AutoTokenizer,
    AutoModel,
    AutoModelForMaskedLM,
    AutoModelForSequenceClassification,
    AutoModelForTokenClassification,
    TrainingArguments,
    Trainer,
    pipeline,
)

#thai2transformers
import thai2transformers
from thai2transformers.preprocess import process_transformers
from thai2transformers.metrics import (
    classification_metrics,
    multilabel_classification_metrics,
)
from thai2transformers.tokenizers import (
    ThaiRobertaTokenizer,
    ThaiWordsNewmmTokenizer,
    ThaiWordsSyllableTokenizer,
    FakeSefrCutTokenizer,
    SEFR_SPLIT_TOKEN
)
```

```python
model_names = [
    'wangchanberta-base-att-spm-uncased',
    'xlm-roberta-base',
    'bert-base-multilingual-cased',
    'wangchanberta-base-wiki-newmm',
    'wangchanberta-base-wiki-ssg',
    'wangchanberta-base-wiki-sefr',
    'wangchanberta-base-wiki-spm',
]

tokenizers = {
    'wangchanberta-base-att-spm-uncased': AutoTokenizer,
    'xlm-roberta-base': AutoTokenizer,
    'bert-base-multilingual-cased': AutoTokenizer,
    'wangchanberta-base-wiki-newmm': ThaiWordsNewmmTokenizer,
    'wangchanberta-base-wiki-ssg': ThaiWordsSyllableTokenizer,
    'wangchanberta-base-wiki-sefr': FakeSefrCutTokenizer,
    'wangchanberta-base-wiki-spm': ThaiRobertaTokenizer,
}
public_models = ['xlm-roberta-base', 'bert-base-multilingual-cased']
#@title Choose Pretrained Model
model_name = "wangchanberta-base-att-spm-uncased" #@param ["wangchanberta-base-att-spm-uncased", "xlm-roberta-base", "bert-base-multilingual-cased", "wangchanberta-base-wiki-newmm", "wangchanberta-base-wiki-syllable", "wangchanberta-base-wiki-sefr", "wangchanberta-base-wiki-spm"]

#create tokenizer
tokenizer = tokenizers[model_name].from_pretrained(
                f'airesearch/{model_name}' if model_name not in public_models else f'{model_name}',
                revision='main',
                model_max_length=416,)
```

```python
#pipeline
fill_mask = pipeline(task='fill-mask',
         tokenizer=tokenizer,
         model = f'airesearch/{model_name}' if model_name not in public_models else f'{model_name}',
         revision = 'main',)
```

```python

preprocess_input_text = True #@param {type:"boolean"}
```

```python
#@title Choose Multi-class Classification Dataset
dataset_name = "wisesight_sentiment" #@param ['wisesight_sentiment','wongnai_reviews']

#pipeline
classify_multiclass = pipeline(task='sentiment-analysis',
         tokenizer=tokenizer,
         model = f'airesearch/{model_name}' if model_name not in public_models else f'{model_name}',
         revision = f'finetuned@{dataset_name}')
```

```python
import re

def extract_agent_speech(text):
    agent_lines = []
    lines = text.split("\n")
    is_agent = False

    for line in lines:
        if re.match(r"\*\*Speaker 1", line):
            is_agent = True
            continue  # Skip the speaker identifier line
        elif re.match(r"\*\*Speaker 2", line):
            is_agent = False
            continue  # Skip the speaker identifier line

        if is_agent:
            agent_lines.append(line.strip())

    return "\n".join(agent_lines)

# Example usage
text = """### บทสนทนาที่ 5
**Speaker 1 (Contact Center Agent):**
1330 สวัสดีครับ ณัฐพล รับสาย ยินดีให้บริการครับ

**Speaker 2 (Citizen):**
สวัสดีครับ พี่อยากสอบถามเกี่ยวกับการลงทะเบียนบัตรทองครับ

**Speaker 1:**
ครับ คุณพี่ต้องการทราบอะไรครับ?

**Speaker 2:**
คือพี่อยากรู้ว่าต้องทำยังไงบ้างในการลงทะเบียนบัตรทองครับ

**Speaker 1:**
เข้าใจครับ คุณพี่ ในการลงทะเบียนบัตรทอง มันต้องเตรียมเอกสารดังนี้ครับ:
1. บัตรประจำตัวประชาชน
2. ทะเบียนบ้าน
3. หนังสือรับรองการย้ายที่อยู่ (ถ้ามี)

**Speaker 2:**
พี่มีบัตรประจำตัวประชาชนและทะเบียนบ้านครับ แล้วต้องไปที่ไหนครับ?

**Speaker 1:**
มันสามารถไปลงทะเบียนได้ที่โรงพยาบาลใกล้บ้านหรือที่สำนักงานหลักประกันสุขภาพแห่งชาติ (สปสช.) ประจำจังหวัดครับ

**Speaker 2:**
พี่ต้องเตรียมอะไรเพิ่มเติมไหมครับ?

**Speaker 1:**
บอกให้เอาเอกสารไปไอเหี้ย แต่ขอแนะนำให้คุณพี่ไปในช่วงเวลาที่ไม่ยุ่งเพื่อหลีกเลี่ยงการรอคอยนานครับ เขามันน่ารัก

**Speaker 2:**
ขอบคุณมากครับ แล้วถ้ามีข้อสงสัยเพิ่มเติม ต้องติดต่อที่ไหนครับ?

**Speaker 1:**
คุณพี่สามารถติดต่อได้ที่สายด่วน 1330 หรือ มันไปดูที่เว็บได้ตลอด 24 ชั่วโมงครับ ยินดีให้บริการครับ

**Speaker 2:**
ขอบคุณครับ

**Speaker 1:**
ยินดีครับ คุณพี่ ขอให้ท่านมีวันที่ดีครับ สวัสดีครับ"""

agent_speech = extract_agent_speech(text)
print(agent_speech)
```

```python
for line in agent_speech.split('\n'):
  if line.strip():  # Skip empty lines
    if preprocess_input_text:
      if model_name not in public_models:
        line = process_transformers(line)

    result = classify_multiclass(line)
    print(f"Line: {line}")
    print(f"Sentiment: {result}")
    print("---") # Separator between lines
```

```python
def check_for_neg(agent_speech):
    for line in agent_speech.split('\n'):
        if line.strip():  # Skip empty lines
            if preprocess_input_text:
                if model_name not in public_models:
                    line = process_transformers(line)

            result = classify_multiclass(line)
            if result[0]['label'] == 'neg':
                return 0  # Found negative sentiment, return 0
    return 1  # No negative sentiment found, return 1

result_bad_word_detect = check_for_neg(agent_speech)

print(f"Final Result: {result_bad_word_detect}")
```
